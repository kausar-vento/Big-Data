```python
import pyspark.sql.functions as F
from pyspark.sql import SparkSession

# Membuat sesi Spark
spark = SparkSession.builder.appName("Uas").getOrCreate()

# Membaca data dari file CSV
df = spark.read.csv('/content/drive/MyDrive/Big Data/data.csv', header=True, inferSchema=True)

# Menambahkan filter tahun
df = df.filter((df['tahun'] >= 2011) & (df['tahun'] <= 2022))

# Melakukan grouping data
df = df.groupBy('nama_kabupaten_kota', 'pendidikan').sum('jumlah_pengangguran')

# Melakukan pivot table
df = df.groupBy('nama_kabupaten_kota').pivot('pendidikan').sum('sum(jumlah_pengangguran)')

# Mengganti nilai null dengan 0
df = df.fillna(0)

# Menampilkan data tertinggi
df = df.withColumn('max_value', F.greatest(*[F.col(col) for col in df.columns[1:]]))
df = df.select('nama_kabupaten_kota', 'max_value')

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

pandas_df = df.toPandas()
sns.barplot(x='nama_kabupaten_kota', y='max_value', data=pandas_df)
plt.xlabel('Kabupaten/Kota')
plt.ylabel('Jumlah Pengangguran Tertinggi')
plt.xticks(rotation=90)
plt.title('Data Pengangguran Tertinggi per Kabupaten/Kota')
plt.show()
```
